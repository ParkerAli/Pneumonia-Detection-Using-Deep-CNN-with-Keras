```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, BatchNormalization
from pathlib import Path
from keras.preprocessing.image import ImageDataGenerator
from keras.models import model_from_json

batch_size = 64
# create generator
train_gen = ImageDataGenerator(horizontal_flip=True)
val_gen = ImageDataGenerator(rescale=1. / 255)
test_gen = ImageDataGenerator()
# prepare an iterators for each dataset
train_datagen = train_gen.flow_from_directory('/content/gdrive/My Drive/Colab Notebooks/chest_xray(GoodSplit)/train/',
                                       target_size=(256, 256),
                                       batch_size= batch_size,
                                       class_mode='binary')
val_datagen = val_gen.flow_from_directory('/content/gdrive/My Drive/Colab Notebooks/chest_xray(GoodSplit)/val/',
                                     target_size=(256, 256),
                                     batch_size= batch_size,
                                     class_mode='binary')

test_datagen = test_gen.flow_from_directory('/content/gdrive/My Drive/Colab Notebooks/chest_xray(GoodSplit)/test/',
                                     target_size=(256, 256),
                                     batch_size= batch_size,
                                     class_mode='binary')
```

### M1

```python
max_acc = 0
for i in range(10):
  print('Iteration', i+1)
  model = Sequential()
  model.add(Conv2D(32,
                  (3, 3),
                  padding = 'same',
                  activation = 'relu',
                  input_shape=(256, 256,3)))

  model.add(Conv2D(64, (3, 3), padding='same', activation="relu"))
  model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
  model.add(MaxPooling2D(pool_size=(2,2)))

  model.add(Dropout(0.25))

  model.add(Conv2D(64, (3, 3), padding='same', activation="relu"))
  model.add(Conv2D(64, (3, 3), padding='same', activation="relu"))
  model.add(MaxPooling2D(pool_size=(2,2)))

  model.add(Flatten())
  model.add(Dense(512, activation="relu"))

  model.add(Dropout(0.50))

  model.add(Dense(1, activation="sigmoid"))

  model.compile(
      loss='binary_crossentropy',
      optimizer='rmsprop', 
      metrics=['accuracy'] 
  )

  model.fit_generator(
        train_datagen,
        steps_per_epoch=5243 // batch_size, 
        epochs = 5,
        validation_data=val_datagen,
        validation_steps= 16 // batch_size)

  acc = model.evaluate_generator(val_datagen)[1]
  print('Iteration accuracy =',acc)
  if(acc > max_acc):
    max_acc = acc
    model.save_weights('/content/gdrive/My Drive/model_weights(M1E5).h5')
    print ("New Maximum =", acc)
```
